```python
import re
import string
import nltk
```

```python
nok_isaretleri=string.punctuation
etkisiz_kelime=open('stopwords/turkce-stop-words.txt','r',encoding='utf-8').read().split()
```

```python
etkisiz_kelime.extend('hakkında eski ilk büyük ardından kişi gelen bulunan nedeniyle yeni yok zaman gün yeni güzel dedi söyledi konuştu ifade büyüka aynı geçen yüzde dr yıl önemli yüksek yıl yer birlikte ayında yılın berabere yenerek son özellikle ortaya fazla iyi son verdi ortaya genel'.split())
```

```python
def veri_temizleme(metin):
    metin=metin.lower()
    metin=metin.replace("\\n"," ")
    metin=re.sub("'(\w+)","",metin)
    metin=re.sub("’(\w+)","",metin)
    metin=re.sub("[“,‛,’,”]","",metin)
    metin=re.sub("[0-9]+","",metin)
    metin="".join(list(map(lambda x:x if x not in nok_isaretleri else " ",metin)))
    metin=" ".join([i for i in metin.split() if i not in etkisiz_kelime])
    metin=" ".join([i for i in metin.split() if len(i) > 1])
    
    return metin
```

```python
db["temiz_metin"]= db["METIN"].apply(lambda x: veri_temizleme(x))
```

```python
db['token']=db["temiz_metin"].apply(lambda x:x.split())
```

```python
db
```

```python
import smart_open
smart_open.open = smart_open.smart_open
import gensim
import pyLDAvis.gensim_models
```

```python
token_metin=db['token']
kelime_liste=gensim.corpora.Dictionary(token_metin)
kelime_liste.filter_extremes(no_below=1,no_above=0.70)
```

```python
dokuman_terim_matris=[kelime_liste.doc2bow(terim) for terim in token_metin]
```

```python
lda_model=gensim.models.ldamodel.LdaModel(corpus=dokuman_terim_matris,id2word=kelime_liste,num_topics=13,passes=10)
```

```python
konular=lda_model.print_topics(num_words=7)
for konu in konular:
    print(konu)
    
    
```

```python
from gensim.models import CoherenceModel

konu_sayisi_aralik_listesi= range(9,40,4)
```

```python
tutarlilik_skorlar_listesi=list()
konu_sayisi_listesi=list()

for konu_sayisi in konu_sayisi_aralik_listesi:
    lda_model = gensim.models.ldamodel.LdaModel(corpus=dokuman_terim_matris,id2word=kelime_liste,num_topics=konu_sayisi,passes=10)
    tutarlilik_model_lda=CoherenceModel(model=lda_model,texts=token_metin,dictionary=kelime_liste,coherence='u_mass')
    gecici_tutarlilik_skoru_lda=tutarlilik_model_lda.get_coherence()
    tutarlilik_skorlar_listesi.append(gecici_tutarlilik_skoru_lda)
    konu_sayisi_listesi.append(konu_sayisi)
```

```python
import matplotlib.pyplot as plt

plt.plot(konu_sayisi_listesi,tutarlilik_skorlar_listesi,"-"),
plt.xlabel('Konu Sayilari')
plt.ylabel('Tutarlilik Skorlari')
plt.show()
```
